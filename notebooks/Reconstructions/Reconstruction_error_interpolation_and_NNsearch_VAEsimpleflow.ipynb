```python
generator_path   = os.path.join(params['module_dir'],'decoder')
encoder_path     = os.path.join(params['module_dir'],'encoder')
nvp_path         = os.path.join(params['module_dir'],'flow')
```

```python
shape = params['data_shape']
shape = np.append(-1,shape)
print(shape)

x_valid = np.reshape(x_valid, shape)
```

```python
import tensorflow.compat.v1 as tf
import tensorflow_probability as tfp
import tensorflow_hub as hub
tfd = tfp.distributions
tfb = tfp.bijectors
```

```python
tf.reset_default_graph()

data          = tf.placeholder(shape=[None]+params['data_shape'],dtype=tf.float32)
latent_point  = tf.placeholder(shape=[None,params['latent_size']],dtype=tf.float32)
u_enc         = tf.placeholder(shape=[None,params['latent_size']],dtype=tf.float32)
sample_size   = tf.placeholder_with_default(params['batch_size'], shape=[])
encoder       = hub.Module(encoder_path, trainable=False)
decoder       = hub.Module(generator_path, trainable=False)
nvp           = hub.Module(nvp_path, trainable=False)
prior         = tfd.MultivariateNormalDiag(tf.zeros(params['latent_size']), scale_identity_multiplier=1.0)

encoded, _    = tf.split(encoder({'x':data},as_dict=True)['z'], 2, axis=-1)
decoded       = decoder({'z':encoded},as_dict=True)['x']

decoded_latent= decoder({'z':latent_point},as_dict=True)['x']

sigma         = tf.reduce_mean(tf.sqrt(tf.square(data-decoded)),axis=0)

samples         = prior.sample(sample_size)
decoded_samples = decoder({'z':samples},as_dict=True)['x']
nvp_samples     = nvp({'z_sample':np.zeros((1,params['latent_size'])),'sample_size':1, 'u_sample':samples}, as_dict=True)['fwd_pass']
u_samples       = nvp({'z_sample':encoded,'sample_size':1, 'u_sample':samples}, as_dict=True)['bwd_pass']
z_samples       = nvp({'z_sample':encoded,'sample_size':1, 'u_sample':u_enc}, as_dict=True)['fwd_pass']
decoded_nvp_samples = decoder({'z':nvp_samples},as_dict=True)['x']
decoded_samples = decoder({'z':z_samples},as_dict=True)['x']
```

```python
sess = tf.Session()
sess.run(tf.global_variables_initializer())
```

```python
from sklearn.neighbors import NearestNeighbors
from sklearn.neighbors import KDTree

np.random.seed(45026)

random_nvp_samples=[]
for ii in range(10000//params['batch_size']+1):
    random_nvp_samples+=[sess.run(decoded_nvp_samples)]
random_nvp_samples=np.asarray(random_nvp_samples).reshape((-1,28,28))[0:10000]

u_encoded_train = []
x_train = x_train.reshape((-1,28,28,1))
for ii in range(60000//params['batch_size']):
    u_encoded_train+=[sess.run(u_samples, feed_dict={data:x_train[ii*params['batch_size']:(ii+1)*params['batch_size']]})]
u_encoded_train   = np.asarray(u_encoded_train).reshape((-1,40))
u_encoded_samples = sess.run(u_samples, feed_dict={data:np.expand_dims(random_nvp_samples[0:params['batch_size']],-1)})

kdt = KDTree(u_encoded_train, leaf_size=10, metric='manhattan')
nns = kdt.query(u_encoded_samples, k=1, return_distance=True)

train_fwd = sess.run(decoded_samples, feed_dict={u_enc:np.squeeze(u_encoded_train[nns[1]])})

plt.scatter(u_encoded_train[nns[1]][0],u_encoded_train[nns[1]][1])
plt.scatter(u_encoded_samples[0],u_encoded_samples[1])

nns = np.squeeze(nns[1])

plt.figure(figsize=(6,6))
plt.subplots_adjust(left=None, bottom=None, right=None, top=None, wspace=0.05, hspace=0.05)
for ii in range(25):
    plt.subplot(5,5,ii+1)
    plt.imshow(np.squeeze((random_nvp_samples[ii]+0.5).reshape(shape)),cmap='gray',vmin=0, vmax=1)
    plt.axis('off')
plt.savefig(os.path.join(params['plot_dir'],'nvp_samples_for_nns.pdf'),bbox_inches='tight')
plt.show()

plt.figure(figsize=(6,6))
plt.subplots_adjust(left=None, bottom=None, right=None, top=None, wspace=0.05, hspace=0.05)
for ii in range(25):
    plt.subplot(5,5,ii+1)
    plt.imshow(np.squeeze((train_fwd[ii]+0.5).reshape(shape)),cmap='gray',vmin=0, vmax=1)
    plt.axis('off')
plt.savefig(os.path.join(params['plot_dir'],'nvp_samples_nns_in_train_model.pdf'),bbox_inches='tight')
plt.show()

plt.figure(figsize=(6,6))
for ii in range(25):
    plt.subplot(5,5,ii+1)
    plt.imshow(np.squeeze((x_train[nns[ii]]+0.5).reshape(shape)),cmap='gray',vmin=0, vmax=1)
    plt.axis('off')
plt.savefig(os.path.join(params['plot_dir'],'nvp_samples_nns_in_train.pdf'),bbox_inches='tight')
plt.show()

```
